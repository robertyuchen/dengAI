```r
# Functions to impute data
replaceNaWithLast <- function(x, a = !is.na(x)) {
    return(x[which(a)[c(1,1:sum(a))][cumsum(a)+1]]);
}
dataImputation <- function(data, cn) {
    for(i in c(1:nrow(data))) {
        if(is.na(data[i,cn])) data[i,cn] <- mean(c(data[i-1,cn],data[i+1,cn]));
    }
    data[,cn] <- replaceNaWithLast(data[,cn]);
    return(data);
}
getMva <- function(data, cn, n  = 30) {
    data[1,  paste('mva_', n, '_', cn, sep  = '')] <- data[1, cn];
    for(i in 2:nrow(data)){
        data[i, paste('mva_', n, '_', cn, sep  = '')] <- mean(data[max(1,(i-n+1)):i, cn]);
    }
    return(data);
}
```

```r
sjAllData <- allData[allData$city == 'sj',];
sjAllData <- sjAllData[order(sjAllData$week_start_date, decreasing = FALSE),];
for(cn in setdiff(colnames(sjAllData), c('city', 'year', 'weekofyear', 'week_start_date', 'total_cases',  'tag'))) {
    sjAllData <- dataImputation(sjAllData, cn);
    sjAllData <- getMva(sjAllData, cn,  5);
    sjAllData <- getMva(sjAllData, cn, 30);
}
plot(as.Date(sjAllData$week_start_date[sjAllData$tag == 0]), sjAllData$total_cases[sjAllData$tag == 0]);
library(MASS);
truehist(sjAllData$total_cases);
```

```r
sjCor <- cor(sjAllData[sjAllData$tag == 0, 
              setdiff(colnames(sjAllData), c('city', 'year', 'week_start_date', 'tag'))]);
sjCor
```

```r
# install.packages('data.table');
library(data.table)
t(sort(sjCor[rownames(sjCor) == 'total_cases',], decreasing  = TRUE));
t(sort(sjCor[rownames(sjCor) == 'total_cases', !(colnames(sjCor)%like%'mva_*')], decreasing  = TRUE));

getPredictorCor  <- function(corMtx, cn) {
    return(t(sort(corMtx[rownames(corMtx) == cn,], decreasing  = TRUE)));
}

getPredictorCor(sjCor,'mva_30_station_max_temp_c');

getPredictors <-  function(corMtx, sortedCor) {
    candidates  <- setdiff(names(sortedCor),  
                           names(sortedCor)[names(sortedCor)%like%'*ndvi*'|names(sortedCor)=='total_cases'|names(sortedCor)=='ws']);
    predictors  <-  candidates[1];
    candidates2 <- setdiff(candidates, candidates[abs(getPredictorCor(corMtx,candidates[1]))>0.5]);
    predictors <- c(predictors,candidates2[1]);
    candidates3 <- setdiff(candidates2, candidates2[abs(getPredictorCor(corMtx,candidates2[1]))>0.5]);
    predictors <- c(predictors, candidates3[1]);
    candidates4 <- setdiff(candidates3, candidates3[abs(getPredictorCor(corMtx,candidates3[1]))>0.5]);
    predictors <- c(predictors, candidates4[length(candidates4)]);
    return(predictors);
}

sjBasePredictors <- getPredictors(sjCor, sort(sjCor[rownames(sjCor) == 'total_cases',], decreasing  = TRUE));
sjBasePredictors <- c('weekofyear', sjBasePredictors);
sjBasePredictors
```

```r
# install.packages('ranger');
library(ranger);
wsTarget <- 'tag';
wsPredictors  <- setdiff(colnames(sjAllData), c('city', 'year', 'weekofyear', 'week_start_date', 'total_cases', 'tag', 'ws'));
trainWsRf <- function(data) {
    fmla <- as.formula(paste(wsTarget,'~', paste(wsPredictors, collapse = '+', sep = ''), sep = ''));
    rf <- ranger(formula = fmla, data = data[,c(wsTarget, wsPredictors)], 
                 num.trees = 100, max.depth = 1, importance = 'impurity');
    return(rf);
}
```

```r
sjWsRf <- trainWsRf(sjAllData);
sjAllData$ws  <- predict(sjWsRf,  sjAllData[,wsPredictors])$predictions;
plot(as.Date(sjAllData$week_start_date), sjAllData$ws);
t(sort(importance(sjWsRf), decreasing = TRUE));
t(quantile(sjAllData$ws[sjAllData$tag == 0], c(1:9)/10));
```

```r
# Train base model
```

```r
evalPredictions <- function(Ya, Yp) return(sum(abs(Ya-Yp))/length(Ya));
mva <- function(x,n) {
  rs = x;
  for(i in n:length(x)){
    rs[i] = mean(x[(i-n+1):i]);
  }
  return(rs);
}
```

```r
# xgb
mva <- function(x,n) {
  rs = x;
  for(i in n:length(x)){
    rs[i] = mean(x[(i-n+1):i]);
  }
  return(rs);
}

# install.packages('xgboost');
library(xgboost)
trainXgb <- function(data, target, predictors) {
  xgb <- xgboost(data = as.matrix(data[,predictors]), label  = data[,target], nrounds = 5000,  
                 eta = 0.01, subsample = 0.7,  verbose = FALSE);
  return(xgb);
}

sjBasePredictors <- sjBasePredictors[!is.na(sjBasePredictors)] #<- c('weekofyear','mva_30_station_max_temp_c','mva_30_station_precip_mm');
iqBasePredictors #<- c('mva_5_reanalysis_min_air_temp_k','mva_30_reanalysis_tdtr_k','mva_30_station_min_temp_c');
target <- 'total_cases';

sjXgb <- trainXgb(sjTrainData,target,sjBasePredictors);
iqXgb <- trainXgb(iqTrainData,target,iqBasePredictors);

sjTrainData$base <- mva(predict(sjXgb, as.matrix(sjTrainData[,sjBasePredictors])),5);
sjTestData$base  <-  predict(sjXgb,  as.matrix(sjTestData[,sjBasePredictors]));
evalPredictions(sjTestData$total_cases,sjTestData$base)

iqTrainData$base <- mva(predict(iqXgb, as.matrix(iqTrainData[,iqBasePredictors])),5);
iqTestData$base  <-  predict(iqXgb,  as.matrix(iqTestData[,iqBasePredictors]));
evalPredictions(iqTestData$total_cases,iqTestData$base)
```

```r
sjSpecPredictors <- c('base', sjSpecPredictors);#<- c('mva_30_station_max_temp_c','mva_5_reanalysis_specific_humidity_g_per_kg',
                      #'mva_5_reanalysis_dew_point_temp_k');
iqSpecPredictors <- c('base', iqSpecPredictors);#<- c('mva_5_reanalysis_precip_amt_kg_per_m2', 'mva_5_reanalysis_specific_humidity_g_per_kg',
                      #'mva_5_reanalysis_dew_point_temp_k');
target <- 'total_cases';


sjSpecRf <- trainXgb(sjTrainData, target,sjSpecPredictors);
iqSpecRf <- trainXgb(iqTrainData, target,iqSpecPredictors);

sjTestData$spec  <-  predict(sjSpecRf,  as.matrix(sjTestData[,sjSpecPredictors]));
sjTestData$pred  <- mva(sjTestData$spec*sjTestData$ws+sjTestData$base*(1-sjTestData$ws), 5);
plot(sjTestData$pred,  type = 'l')
evalPredictions(sjTestData$total_cases,sjTestData$pred)



#iqTestData$spec <-  predict(iqSpecRf,  iqTestData[,iqSpecPredictors])$predictions;
iqTestData$spec <- predict(iqSpecRf, as.matrix(iqTestData[,iqSpecPredictors]));
iqTestData$pred  <- mva(iqTestData$spec*iqTestData$ws+iqTestData$base*(1-iqTestData$ws), 5);
plot(iqTestData$pred, type = 'l')
evalPredictions(iqTestData$total_cases,iqTestData$pred)
```
